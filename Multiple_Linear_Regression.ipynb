```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel("/content/age_experience_salary_dataset.xlsx")
dataset.head(4)
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
sns.pairplot(data =dataset)
plt.show
```

```python
sns.heatmap(data = dataset.corr(),annot=True)
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Salary"]
```

```python
dataset.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
# y = m1x1+m2x2+c formula of multiple linear equation.
```

```python
lr.coef_
```

```python
lr.intercept_
```
