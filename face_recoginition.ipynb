```python
<!DOCTYPE html>
<html lang="en">
<head>
<meta charset="UTF-8">
<meta name="viewport" content="width=device-width, initial-scale=1.0">
<title>VISION AI ‚Äî Object Detection & Tracker</title>
<link href="https://fonts.googleapis.com/css2?family=Share+Tech+Mono&family=Orbitron:wght@700;900&display=swap" rel="stylesheet">
<style><!DOCTYPE html>
<html lang="en">
<head>
<meta charset="UTF-8">
<meta name="viewport" content="width=device-width, initial-scale=1.0">
<title>VISION AI ‚Äî Object Detection & Tracker</title>
<link href="https://fonts.googleapis.com/css2?family=Share+Tech+Mono&family=Orbitron:wght@700;900&display=swap" rel="stylesheet">
<style>
:root{
  --bg:#03080f;--panel:#070f1a;--accent:#00f5ff;--accent2:#ff006e;
  --green:#39ff14;--warn:#ffb800;--dim:#3a6070;--text:#b0d8e8;
  --border:rgba(0,245,255,0.18);
}
*{margin:0;padding:0;box-sizing:border-box}
body{background:var(--bg);color:var(--text);font-family:'Share Tech Mono',monospace;min-height:100vh;
  background-image:linear-gradient(rgba(0,245,255,0.025) 1px,transparent 1px),linear-gradient(90deg,rgba(0,245,255,0.025) 1px,transparent 1px);
  background-size:36px 36px}
body::after{content:'';position:fixed;inset:0;background:repeating-linear-gradient(0deg,transparent,transparent 3px,rgba(0,0,0,0.06) 3px,rgba(0,0,0,0.06) 4px);pointer-events:none;z-index:9999}

/* HEADER */
header{display:flex;align-items:center;justify-content:space-between;padding:12px 24px;
  border-bottom:1px solid var(--border);background:rgba(3,8,15,0.95);backdrop-filter:blur(8px);
  position:sticky;top:0;z-index:100}
.logo{font-family:'Orbitron',sans-serif;font-size:1.3rem;font-weight:900;color:var(--accent);
  text-shadow:0 0 18px var(--accent),0 0 36px rgba(0,245,255,0.4);letter-spacing:5px}
.logo em{color:var(--accent2);font-style:normal}
.hdr-status{display:flex;gap:18px;font-size:0.68rem;color:var(--dim)}
.st{display:flex;align-items:center;gap:6px}
.dot{width:7px;height:7px;border-radius:50%;background:var(--dim);transition:.3s}
.dot.on{background:var(--green);box-shadow:0 0 8px var(--green);animation:blink 1.5s infinite}
.dot.err{background:var(--accent2);box-shadow:0 0 8px var(--accent2)}
.dot.warn{background:var(--warn);box-shadow:0 0 8px var(--warn)}
@keyframes blink{0%,100%{opacity:1}50%{opacity:.35}}

/* LAYOUT */
.layout{display:grid;grid-template-columns:1fr 330px;gap:14px;padding:14px;max-width:1350px;margin:0 auto}

/* CAMERA */
.cam-wrap{position:relative;border:1px solid var(--border);border-radius:4px;background:#000;
  overflow:hidden;aspect-ratio:16/9;min-height:400px}
.cam-wrap::before{content:'';position:absolute;inset:0;
  background:linear-gradient(135deg,rgba(0,245,255,0.03) 0%,transparent 50%,rgba(255,0,110,0.02) 100%);
  pointer-events:none;z-index:2}
.corner{position:absolute;width:18px;height:18px;z-index:5}
.tl{top:8px;left:8px;border-top:2px solid var(--accent);border-left:2px solid var(--accent)}
.tr{top:8px;right:8px;border-top:2px solid var(--accent);border-right:2px solid var(--accent)}
.bl{bottom:8px;left:8px;border-bottom:2px solid var(--accent);border-left:2px solid var(--accent)}
.br{bottom:8px;right:8px;border-bottom:2px solid var(--accent);border-right:2px solid var(--accent)}
#vid{width:100%;height:100%;object-fit:cover;display:block;transform:scaleX(-1)}
#cvs{position:absolute;top:0;left:0;width:100%;height:100%;pointer-events:none;z-index:3;transform:scaleX(-1)}
#fps{position:absolute;top:12px;right:12px;z-index:5;font-size:.63rem;color:var(--green);
  background:rgba(0,0,0,.65);padding:3px 8px;border:1px solid rgba(57,255,20,.25);border-radius:2px}
.live-tag{position:absolute;top:12px;left:12px;z-index:5;font-family:'Orbitron',sans-serif;
  font-size:.55rem;color:var(--accent2);background:rgba(0,0,0,.65);padding:3px 8px;
  border:1px solid rgba(255,0,110,.25);border-radius:2px;display:flex;align-items:center;gap:5px;letter-spacing:2px}
.rec{width:6px;height:6px;border-radius:50%;background:var(--accent2);animation:blink .8s infinite}

/* LOADING */
#loader{position:absolute;inset:0;background:rgba(3,8,15,.95);display:flex;flex-direction:column;
  align-items:center;justify-content:center;gap:20px;z-index:10}
.ld-title{font-family:'Orbitron',sans-serif;font-size:1rem;color:var(--accent);
  text-shadow:0 0 16px var(--accent);letter-spacing:4px}
.ld-track{width:280px;height:3px;background:rgba(0,245,255,.1);border-radius:2px;overflow:hidden;border:1px solid var(--border)}
.ld-bar{height:100%;background:var(--accent);box-shadow:0 0 8px var(--accent);width:0%;transition:width .4s}
.ld-msg{font-size:.65rem;color:var(--dim);text-align:center;max-width:280px;line-height:1.9}

/* BUTTONS */
.cam-btns{position:absolute;bottom:14px;left:50%;transform:translateX(-50%);z-index:5;display:flex;gap:10px}
.btn{padding:9px 20px;border:1px solid var(--accent);background:rgba(0,245,255,.07);color:var(--accent);
  font-family:'Orbitron',sans-serif;font-size:.6rem;font-weight:700;letter-spacing:2px;cursor:pointer;
  border-radius:2px;text-transform:uppercase;transition:.2s}
.btn:hover{background:rgba(0,245,255,.18);box-shadow:0 0 16px rgba(0,245,255,.3)}
.btn.red{border-color:var(--accent2);color:var(--accent2);background:rgba(255,0,110,.07)}
.btn.red:hover{background:rgba(255,0,110,.18);box-shadow:0 0 16px rgba(255,0,110,.3)}

/* SIDE PANELS */
.side{display:flex;flex-direction:column;gap:14px}
.panel{border:1px solid var(--border);border-radius:4px;background:var(--panel);overflow:hidden}
.ph{padding:9px 14px;background:rgba(0,245,255,.04);border-bottom:1px solid var(--border);
  font-family:'Orbitron',sans-serif;font-size:.6rem;font-weight:700;letter-spacing:3px;color:var(--accent)}

/* MOOD */
.mood-body{padding:16px;display:flex;flex-direction:column;align-items:center;gap:12px}
.mood-emoji{font-size:3.6rem;animation:float 3s ease-in-out infinite;filter:drop-shadow(0 0 12px currentColor)}
@keyframes float{0%,100%{transform:translateY(0)}50%{transform:translateY(-6px)}}
.mood-name{font-family:'Orbitron',sans-serif;font-size:1.15rem;font-weight:900;letter-spacing:3px;transition:.4s}
.mood-sub{font-size:.62rem;color:var(--dim);text-align:center;line-height:1.6;min-height:36px}
.mood-bars{width:100%}
.mb-row{display:flex;align-items:center;gap:8px;margin-bottom:5px;font-size:.64rem}
.mb-lbl{width:72px;color:var(--dim);text-transform:capitalize}
.mb-track{flex:1;height:5px;background:rgba(255,255,255,.05);border-radius:3px;overflow:hidden}
.mb-fill{height:100%;border-radius:3px;transition:width .35s ease}
.mb-pct{width:32px;text-align:right;font-size:.6rem}

/* STATS */
.sg{display:grid;grid-template-columns:1fr 1fr;gap:8px;padding:12px}
.sb{border:1px solid var(--border);border-radius:3px;padding:9px;text-align:center;background:rgba(0,245,255,.02)}
.sv{font-family:'Orbitron',sans-serif;font-size:1.3rem;font-weight:900;color:var(--accent);text-shadow:0 0 8px var(--accent)}
.sl{font-size:.58rem;color:var(--dim);margin-top:3px;letter-spacing:1px}

/* OBJECTS LIST */
.obj-list{padding:10px;overflow-y:auto;max-height:300px}
.oi{display:flex;align-items:center;gap:9px;padding:7px 9px;border:1px solid rgba(0,245,255,.08);
  border-radius:3px;margin-bottom:6px;background:rgba(0,245,255,.025);animation:fdin .25s ease}
@keyframes fdin{from{opacity:0;transform:translateX(8px)}to{opacity:1;transform:none}}
.oi:hover{border-color:var(--accent);background:rgba(0,245,255,.06)}
.oi-icon{font-size:1.4rem;flex-shrink:0}
.oi-info{flex:1;min-width:0}
.oi-name{font-family:'Orbitron',sans-serif;font-size:.67rem;font-weight:700;color:var(--accent);
  text-transform:uppercase;letter-spacing:1px}
.oi-desc{font-size:.58rem;color:var(--dim);margin-top:2px;line-height:1.4;
  white-space:nowrap;overflow:hidden;text-overflow:ellipsis}
.oi-conf{font-size:.62rem;color:var(--green);font-weight:700;text-align:right;flex-shrink:0}
.cb{width:38px;height:3px;background:rgba(57,255,20,.12);border-radius:2px;margin-top:3px;overflow:hidden}
.cf{height:100%;background:var(--green);border-radius:2px}
.empty{text-align:center;padding:24px 12px;color:var(--dim);font-size:.67rem;line-height:2}

/* TICKER */
.ticker{grid-column:1/3;border:1px solid var(--border);border-radius:4px;background:var(--panel);
  padding:9px 16px;display:flex;align-items:center;gap:12px;overflow:hidden}
.tk-lbl{font-family:'Orbitron',sans-serif;font-size:.58rem;color:var(--accent2);letter-spacing:2px;white-space:nowrap}
.tk-txt{font-size:.67rem;color:var(--text);white-space:nowrap;overflow:hidden;text-overflow:ellipsis}

::-webkit-scrollbar{width:3px}::-webkit-scrollbar-thumb{background:var(--border);border-radius:2px}
</style>
</head>
<body>

<header>
  <div class="logo">VISION<em>AI</em></div>
  <div class="hdr-status">
    <div class="st"><div class="dot" id="dc"></div><span id="lc">CAM OFFLINE</span></div>
    <div class="st"><div class="dot" id="dm"></div><span id="lm">LOADING MODEL</span></div>
    <div class="st"><div class="dot" id="dt"></div><span id="lt">TRACKING IDLE</span></div>
  </div>
</header>

<div class="layout">

  <!-- CAMERA -->
  <div class="cam-wrap" id="camWrap">
    <div class="corner tl"></div><div class="corner tr"></div>
    <div class="corner bl"></div><div class="corner br"></div>
    <div class="live-tag"><div class="rec"></div>LIVE</div>
    <div id="fps">-- FPS</div>
    <video id="vid" autoplay muted playsinline></video>
    <canvas id="cvs"></canvas>

    <div id="loader">
      <div class="ld-title">‚¨° LOADING VISION AI ‚¨°</div>
      <div class="ld-track"><div class="ld-bar" id="ldbar"></div></div>
      <div class="ld-msg" id="ldmsg">Downloading COCO-SSD neural network...<br>This takes ~10 seconds on first load</div>
    </div>

    <div class="cam-btns">
      <button class="btn" id="btnStart" onclick="startCam()">‚ñ∂ ACTIVATE CAMERA</button>
      <button class="btn red" id="btnStop" onclick="stopCam()" style="display:none">‚ñ† STOP</button>
    </div>
  </div>

  <!-- SIDE -->
  <div class="side">

    <!-- MOOD -->
    <div class="panel">
      <div class="ph">‚¨° MOOD ANALYSIS</div>
      <div class="mood-body">
        <div class="mood-emoji" id="mEmoji">üòê</div>
        <div class="mood-name" id="mName" style="color:var(--dim)">AWAITING FACE</div>
        <div class="mood-sub" id="mSub">Point your face at the camera to detect your emotion in real time.</div>
        <div class="mood-bars" id="mBars"></div>
      </div>
    </div>

    <!-- STATS -->
    <div class="panel">
      <div class="ph">‚¨° LIVE STATS</div>
      <div class="sg">
        <div class="sb"><div class="sv" id="sObj">0</div><div class="sl">OBJECTS</div></div>
        <div class="sb"><div class="sv" id="sFps">0</div><div class="sl">FPS</div></div>
        <div class="sb"><div class="sv" id="sFrames">0</div><div class="sl">FRAMES</div></div>
        <div class="sb"><div class="sv" id="sTracked">0</div><div class="sl">TRACKED</div></div>
      </div>
    </div>

    <!-- OBJECTS -->
    <div class="panel" style="flex:1">
      <div class="ph">‚¨° DETECTED OBJECTS</div>
      <div class="obj-list" id="objList">
        <div class="empty">Press ACTIVATE CAMERA<br>to start detection</div>
      </div>
    </div>

  </div>

  <!-- TICKER -->
  <div class="ticker">
    <span class="tk-lbl">‚ñ∂ INFO</span>
    <span class="tk-txt" id="tkTxt">VisionAI ready ‚Äî click Activate Camera to begin real-time detection and mood analysis.</span>
  </div>

</div>

<script src="https://cdn.jsdelivr.net/npm/@tensorflow/tfjs@4.10.0/dist/tf.min.js"></script>
<script src="https://cdn.jsdelivr.net/npm/@tensorflow-models/coco-ssd@2.2.2/dist/coco-ssd.min.js"></script>
<script>
// ‚îÄ‚îÄ OBJECT METADATA ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
const OBJ = {
  person:     {i:'üßë',d:'Human detected. Tracking body position.',c:'#00f5ff'},
  cat:        {i:'üê±',d:'Domestic cat. Felis catus.',c:'#ff9f43'},
  dog:        {i:'üê∂',d:'Domestic dog. Canis lupus familiaris.',c:'#ff9f43'},
  car:        {i:'üöó',d:'Automobile vehicle.',c:'#54a0ff'},
  truck:      {i:'üöõ',d:'Large motor truck.',c:'#54a0ff'},
  bus:        {i:'üöå',d:'Public transport bus.',c:'#54a0ff'},
  bicycle:    {i:'üö≤',d:'Human-powered two-wheel vehicle.',c:'#00b894'},
  motorcycle: {i:'üèçÔ∏è',d:'Two-wheeled motor vehicle.',c:'#00b894'},
  chair:      {i:'ü™ë',d:'Seating furniture.',c:'#a29bfe'},
  couch:      {i:'üõãÔ∏è',d:'Upholstered sofa for sitting.',c:'#a29bfe'},
  bottle:     {i:'üç∂',d:'Container for liquids.',c:'#55efc4'},
  cup:        {i:'‚òï',d:'Small cup for beverages.',c:'#ffeaa7'},
  bowl:       {i:'ü•£',d:'Round open container for food.',c:'#ffeaa7'},
  laptop:     {i:'üíª',d:'Portable personal computer.',c:'#74b9ff'},
  keyboard:   {i:'‚å®Ô∏è',d:'Input device for typing.',c:'#b2bec3'},
  mouse:      {i:'üñ±Ô∏è',d:'Computer pointing device.',c:'#b2bec3'},
  tv:         {i:'üì∫',d:'Television display screen.',c:'#6c5ce7'},
  cell_phone: {i:'üì±',d:'Mobile phone device.',c:'#fd79a8'},
  book:       {i:'üìö',d:'Printed or written document.',c:'#fdcb6e'},
  clock:      {i:'üïê',d:'Timekeeping device.',c:'#e17055'},
  scissors:   {i:'‚úÇÔ∏è',d:'Cutting tool with two blades.',c:'#636e72'},
  backpack:   {i:'üéí',d:'Bag worn on the back.',c:'#fdcb6e'},
  handbag:    {i:'üëú',d:'Small handheld bag.',c:'#fd79a8'},
  umbrella:   {i:'‚òÇÔ∏è',d:'Portable rain protection device.',c:'#74b9ff'},
  tie:        {i:'üëî',d:'Formal neckwear garment.',c:'#6c5ce7'},
  sports_ball:{i:'‚öΩ',d:'Spherical sports ball.',c:'#e17055'},
  kite:       {i:'ü™Å',d:'Flying object on a string.',c:'#00b894'},
  pizza:      {i:'üçï',d:'Italian flatbread dish.',c:'#ff9f43'},
  apple:      {i:'üçé',d:'Red fruit from apple tree.',c:'#e17055'},
  orange:     {i:'üçä',d:'Citrus fruit.',c:'#ff9f43'},
  banana:     {i:'üçå',d:'Yellow curved fruit.',c:'#ffeaa7'},
  sandwich:   {i:'ü•™',d:'Bread with filling.',c:'#fdcb6e'},
  cake:       {i:'üéÇ',d:'Baked sweet dessert.',c:'#fd79a8'},
  chair_bed:  {i:'üõèÔ∏è',d:'Furniture for sleeping.',c:'#a29bfe'},
  vase:       {i:'üè∫',d:'Decorative container.',c:'#55efc4'},
  toothbrush: {i:'ü™•',d:'Dental hygiene brush.',c:'#00b894'},
  _default:   {i:'üì¶',d:'Unclassified object tracked.',c:'#a29bfe'}
};

// ‚îÄ‚îÄ MOOD CONFIG ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
const MOODS = {
  happy:    {e:'üòÑ',c:'#f9ca24',d:'Smile and raised cheeks detected. Positive emotion!'},
  sad:      {e:'üò¢',c:'#74b9ff',d:'Downcast expression. Low energy facial cues.'},
  angry:    {e:'üò°',c:'#ff7675',d:'Tense jaw and furrowed brow detected.'},
  surprised:{e:'üò≤',c:'#fdcb6e',d:'Wide eyes and open mouth detected.'},
  fearful:  {e:'üò®',c:'#a29bfe',d:'Raised brow, widened eyes detected.'},
  neutral:  {e:'üòê',c:'#b2bec3',d:'Relaxed muscles. No dominant emotion.'},
  disgusted:{e:'ü§¢',c:'#55efc4',d:'Nose wrinkle, upper lip raised.'}
};

// ‚îÄ‚îÄ STATE ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
let model = null, stream = null, running = false;
let frameCount = 0, fps = 0, lastFpsT = Date.now(), totalFrames = 0, totalTracked = 0;
let trackerMap = {}, trackCnt = 0;
let detectEvery = 3, frameN = 0; // skip frames for speed
let lastPreds = [], lastMood = null;
let moodFrameCount = 0;
let raf = null;

// ‚îÄ‚îÄ MODEL LOAD ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
(async () => {
  const bar = id('ldbar'), msg = id('ldmsg');
  setDot('dm','warn'); setLabel('lm','LOADING...');
  bar.style.width = '15%';
  msg.innerHTML = 'Downloading COCO-SSD model (~5MB)...<br><span style="color:var(--warn)">Please wait, first load takes ~10s</span>';

  try {
    // Use backend=cpu for stability if webgl fails
    try { await tf.setBackend('webgl'); } catch(e) { await tf.setBackend('cpu'); }
    bar.style.width = '40%';
    msg.innerHTML = 'TensorFlow ready. Loading detection model...';

    model = await cocoSsd.load({ base: 'lite_mobilenet_v2' }); // fastest variant
    bar.style.width = '100%';
    msg.innerHTML = '‚úì Model loaded! Click Activate Camera.';
    setDot('dm','on'); setLabel('lm','MODEL READY');
    await sleep(600);
    id('loader').style.display = 'none';
  } catch(e) {
    bar.style.width='100%'; bar.style.background='var(--accent2)';
    msg.innerHTML = '‚úó Load failed: ' + e.message + '<br>Check internet connection.';
    setDot('dm','err'); setLabel('lm','LOAD ERROR');
  }
})();

// ‚îÄ‚îÄ CAMERA ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
async function startCam() {
  if (!model) { tick('‚ö† Model still loading, please wait a moment...'); return; }
  try {
    stream = await navigator.mediaDevices.getUserMedia({video:{width:640,height:480,facingMode:'user'},audio:false});
    const vid = id('vid'), cvs = id('cvs');
    vid.srcObject = stream;
    await vid.play();
    vid.addEventListener('loadedmetadata',()=>{
      cvs.width = vid.videoWidth; cvs.height = vid.videoHeight;
      running = true;
      setDot('dc','on'); setLabel('lc','CAM ACTIVE');
      setDot('dt','on'); setLabel('lt','TRACKING ON');
      id('btnStart').style.display='none'; id('btnStop').style.display='block';
      loop();
    });
  } catch(e) {
    setDot('dc','err'); setLabel('lc','CAM DENIED');
    tick('Camera access denied. Allow camera in browser settings, then retry.');
  }
}

function stopCam() {
  running = false;
  if (raf) cancelAnimationFrame(raf);
  if (stream) stream.getTracks().forEach(t=>t.stop());
  id('vid').srcObject = null;
  const cvs = id('cvs'); cvs.getContext('2d').clearRect(0,0,cvs.width,cvs.height);
  setDot('dc',''); setLabel('lc','CAM OFFLINE');
  setDot('dt',''); setLabel('lt','TRACKING IDLE');
  id('btnStart').style.display='block'; id('btnStop').style.display='none';
  id('objList').innerHTML='<div class="empty">Camera stopped.</div>';
  moodIdle(); trackerMap={}; lastPreds=[];
  id('sObj').textContent='0';
}

// ‚îÄ‚îÄ MAIN LOOP ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
async function loop() {
  if (!running) return;

  const vid = id('vid'), cvs = id('cvs'), ctx = cvs.getContext('2d');

  // FPS counter
  frameCount++; totalFrames++;
  const now = Date.now();
  if (now - lastFpsT >= 1000) {
    fps = frameCount; frameCount = 0; lastFpsT = now;
    id('fps').textContent = fps + ' FPS';
    id('sFps').textContent = fps;
  }

  // Run detection every N frames to keep it smooth
  frameN++;
  if (frameN % detectEvery === 0 && vid.readyState === 4) {
    try {
      lastPreds = await model.detect(vid, 10); // max 10 detections
    } catch(e) { lastPreds = []; }
    updateTracker(lastPreds);
    updateObjectsUI(lastPreds);
    id('sObj').textContent = lastPreds.length;
    id('sFrames').textContent = totalFrames;
    id('sTracked').textContent = totalTracked;

    // Mood from pixel analysis every 10 frames
    if (frameN % 10 === 0) analyzeMoodFromPixels(vid);
  }

  // Draw every frame (smooth)
  ctx.clearRect(0,0,cvs.width,cvs.height);
  drawDetections(ctx, lastPreds, cvs.width, cvs.height);

  raf = requestAnimationFrame(loop);
}

// ‚îÄ‚îÄ TRACKER ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
function updateTracker(preds) {
  const next = {};
  for (const p of preds) {
    const [x,y,w,h] = p.bbox;
    const cx = x+w/2, cy = y+h/2;
    let bestId = null, bestDist = 100;
    for (const [id2, obj] of Object.entries(trackerMap)) {
      if (obj.cls !== p.class) continue;
      const d = Math.hypot(cx-obj.cx, cy-obj.cy);
      if (d < bestDist) { bestDist=d; bestId=id2; }
    }
    if (bestId) {
      next[bestId] = {cls:p.class, cx, cy, age:(trackerMap[bestId].age||0)+1};
    } else {
      const nid = ++trackCnt; totalTracked++;
      next[nid] = {cls:p.class, cx, cy, age:1};
    }
    p._tid = bestId || trackCnt;
  }
  trackerMap = next;
}

// ‚îÄ‚îÄ DRAW ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
function drawDetections(ctx, preds, W, H) {
  for (const p of preds) {
    const info = OBJ[p.class] || OBJ._default;
    let [x,y,w,h] = p.bbox;
    // clamp
    x=Math.max(0,x); y=Math.max(0,y);
    w=Math.min(w,W-x); h=Math.min(h,H-y);
    const conf = Math.round(p.score*100);
    const col = info.c;
    const tid = p._tid || '';

    // Glow box
    ctx.shadowColor=col; ctx.shadowBlur=10;
    ctx.strokeStyle=col; ctx.lineWidth=1.5;
    ctx.strokeRect(x,y,w,h);
    ctx.shadowBlur=0;

    // Corner marks
    const cs=10; ctx.strokeStyle='#fff'; ctx.lineWidth=2;
    for(const[cx2,cy2,dx,dy] of [[x,y,1,1],[x+w,y,-1,1],[x,y+h,1,-1],[x+w,y+h,-1,-1]]){
      ctx.beginPath();ctx.moveTo(cx2+dx*cs,cy2);ctx.lineTo(cx2,cy2);ctx.lineTo(cx2,cy2+dy*cs);ctx.stroke();
    }

    // Label
    const lbl = `${p.class.replace('_',' ').toUpperCase()}  #${tid}  ${conf}%`;
    ctx.font = 'bold 11px "Share Tech Mono",monospace';
    const tw = ctx.measureText(lbl).width;
    const lx = x, ly = y>20 ? y-20 : y+h+2;
    ctx.fillStyle='rgba(0,0,0,.78)';
    ctx.fillRect(lx,ly,tw+12,17);
    ctx.fillStyle=col;
    ctx.fillText(lbl,lx+5,ly+12);

    // Confidence bar at bottom of box
    ctx.fillStyle='rgba(0,0,0,.45)'; ctx.fillRect(x,y+h-4,w,4);
    ctx.fillStyle=col; ctx.fillRect(x,y+h-4,w*p.score,4);
  }
}

// ‚îÄ‚îÄ MOOD FROM PIXEL ANALYSIS ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
// Uses a hidden canvas to sample face-region pixels and infer rough mood
// Not a real model ‚Äî uses brightness, contrast, movement heuristics for fun
let prevBrightness = 0, moodHistory = [];
function analyzeMoodFromPixels(vid) {
  try {
    const tmpC = document.createElement('canvas');
    tmpC.width=80; tmpC.height=60;
    const tc = tmpC.getContext('2d');
    tc.drawImage(vid,0,0,80,60);
    const px = tc.getImageData(0,0,80,60).data;

    let r=0,g=0,b=0,n=px.length/4;
    for(let i=0;i<px.length;i+=4){r+=px[i];g+=px[i+1];b+=px[i+2];}
    r/=n; g/=n; b/=n;
    const brightness = (r+g+b)/3;
    const warmth = r - b; // warm=positive, cool=negative
    const delta = Math.abs(brightness - prevBrightness);
    prevBrightness = brightness;

    // Heuristic mood scores (fun approximation)
    let scores = {
      happy: 0, sad: 0, angry: 0, surprised: 0,
      fearful: 0, neutral: 0, disgusted: 0
    };

    // Brighter frame ‚Üí more positive
    if (brightness > 160) scores.happy += 0.4;
    else if (brightness < 80) scores.sad += 0.35;

    // Warm tones ‚Üí happy/angry
    if (warmth > 20) scores.happy += 0.25;
    else if (warmth < -15) scores.sad += 0.2;

    // High movement ‚Üí surprised/fearful
    if (delta > 18) scores.surprised += 0.4;
    else if (delta > 10) scores.fearful += 0.2;

    // Normalize & add neutral
    const total = Object.values(scores).reduce((a,b)=>a+b,0);
    if (total < 0.3) scores.neutral = 0.7;
    else scores.neutral += Math.max(0, 0.3 - total*0.1);

    // Smooth with history
    moodHistory.push(scores);
    if (moodHistory.length > 5) moodHistory.shift();
    const avg = {};
    for (const k of Object.keys(scores)) {
      avg[k] = moodHistory.reduce((s,m)=>s+(m[k]||0),0)/moodHistory.length;
    }

    // Normalize
    const sum2 = Object.values(avg).reduce((a,b)=>a+b,0)||1;
    for(const k of Object.keys(avg)) avg[k] = avg[k]/sum2;

    updateMoodUI(avg);
  } catch(e) {}
}

function updateMoodUI(scores) {
  const sorted = Object.entries(scores).sort((a,b)=>b[1]-a[1]);
  const [topName, topVal] = sorted[0];
  const cfg = MOODS[topName] || MOODS.neutral;

  id('mEmoji').textContent = cfg.e;
  id('mName').textContent = topName.toUpperCase();
  id('mName').style.color = cfg.c;
  id('mSub').textContent = cfg.d;

  id('mBars').innerHTML = sorted.map(([n,v])=>{
    const mc = (MOODS[n]||MOODS.neutral).c;
    return `<div class="mb-row">
      <div class="mb-lbl">${n}</div>
      <div class="mb-track"><div class="mb-fill" style="width:${Math.round(v*100)}%;background:${mc}"></div></div>
      <div class="mb-pct" style="color:${mc}">${Math.round(v*100)}%</div>
    </div>`;
  }).join('');

  tick(`Mood: ${topName.toUpperCase()} (${Math.round(topVal*100)}%) ‚Äî ${cfg.d}`);
}

function moodIdle() {
  id('mEmoji').textContent='üòê';
  id('mName').textContent='NO FEED';
  id('mName').style.color='var(--dim)';
  id('mSub').textContent='Activate camera to analyze your mood in real time.';
  id('mBars').innerHTML='';
}

// ‚îÄ‚îÄ OBJECTS LIST UI ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
function updateObjectsUI(preds) {
  const el = id('objList');
  if (!preds.length) {
    el.innerHTML='<div class="empty">No objects detected.<br>Move something into frame!</div>';
    return;
  }
  // Deduplicate by class, keep highest confidence
  const uniq = {};
  for(const p of preds) if(!uniq[p.class]||p.score>uniq[p.class].score) uniq[p.class]=p;

  el.innerHTML = Object.values(uniq).sort((a,b)=>b.score-a.score).map(p=>{
    const info = OBJ[p.class]||OBJ._default;
    const conf = Math.round(p.score*100);
    return `<div class="oi">
      <div class="oi-icon">${info.i}</div>
      <div class="oi-info">
        <div class="oi-name">${p.class.replace(/_/g,' ')}</div>
        <div class="oi-desc">${info.d}</div>
      </div>
      <div style="flex-shrink:0;text-align:right">
        <div class="oi-conf">${conf}%</div>
        <div class="cb"><div class="cf" style="width:${conf}%"></div></div>
      </div>
    </div>`;
  }).join('');

  const names = Object.keys(uniq).map(n=>n.replace(/_/g,' ').toUpperCase()).join(', ');
  tick(`Detected: ${names} ‚Äî ${preds.length} object(s) actively tracked.`);
}

// ‚îÄ‚îÄ HELPERS ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
const id = s => document.getElementById(s);
const sleep = ms => new Promise(r=>setTimeout(r,ms));
function tick(msg) { id('tkTxt').textContent = msg; }
function setDot(elId, state) {
  const d = id(elId); d.className='dot';
  if(state==='on') d.className='dot on';
  else if(state==='err') d.className='dot err';
  else if(state==='warn') d.className='dot warn';
}
function setLabel(elId, txt) { id(elId).textContent = txt; }
</script>
</body>
</html>

:root{
  --bg:#03080f;--panel:#070f1a;--accent:#00f5ff;--accent2:#ff006e;
  --green:#39ff14;--warn:#ffb800;--dim:#3a6070;--text:#b0d8e8;
  --border:rgba(0,245,255,0.18);
}
*{margin:0;padding:0;box-sizing:border-box}
body{background:var(--bg);color:var(--text);font-family:'Share Tech Mono',monospace;min-height:100vh;
  background-image:linear-gradient(rgba(0,245,255,0.025) 1px,transparent 1px),linear-gradient(90deg,rgba(0,245,255,0.025) 1px,transparent 1px);
  background-size:36px 36px}
body::after{content:'';position:fixed;inset:0;background:repeating-linear-gradient(0deg,transparent,transparent 3px,rgba(0,0,0,0.06) 3px,rgba(0,0,0,0.06) 4px);pointer-events:none;z-index:9999}

/* HEADER */
header{display:flex;align-items:center;justify-content:space-between;padding:12px 24px;
  border-bottom:1px solid var(--border);background:rgba(3,8,15,0.95);backdrop-filter:blur(8px);
  position:sticky;top:0;z-index:100}
.logo{font-family:'Orbitron',sans-serif;font-size:1.3rem;font-weight:900;color:var(--accent);
  text-shadow:0 0 18px var(--accent),0 0 36px rgba(0,245,255,0.4);letter-spacing:5px}
.logo em{color:var(--accent2);font-style:normal}
.hdr-status{display:flex;gap:18px;font-size:0.68rem;color:var(--dim)}
.st{display:flex;align-items:center;gap:6px}
.dot{width:7px;height:7px;border-radius:50%;background:var(--dim);transition:.3s}
.dot.on{background:var(--green);box-shadow:0 0 8px var(--green);animation:blink 1.5s infinite}
.dot.err{background:var(--accent2);box-shadow:0 0 8px var(--accent2)}
.dot.warn{background:var(--warn);box-shadow:0 0 8px var(--warn)}
@keyframes blink{0%,100%{opacity:1}50%{opacity:.35}}

/* LAYOUT */
.layout{display:grid;grid-template-columns:1fr 330px;gap:14px;padding:14px;max-width:1350px;margin:0 auto}

/* CAMERA */
.cam-wrap{position:relative;border:1px solid var(--border);border-radius:4px;background:#000;
  overflow:hidden;aspect-ratio:16/9;min-height:400px}
.cam-wrap::before{content:'';position:absolute;inset:0;
  background:linear-gradient(135deg,rgba(0,245,255,0.03) 0%,transparent 50%,rgba(255,0,110,0.02) 100%);
  pointer-events:none;z-index:2}
.corner{position:absolute;width:18px;height:18px;z-index:5}
.tl{top:8px;left:8px;border-top:2px solid var(--accent);border-left:2px solid var(--accent)}
.tr{top:8px;right:8px;border-top:2px solid var(--accent);border-right:2px solid var(--accent)}
.bl{bottom:8px;left:8px;border-bottom:2px solid var(--accent);border-left:2px solid var(--accent)}
.br{bottom:8px;right:8px;border-bottom:2px solid var(--accent);border-right:2px solid var(--accent)}
#vid{width:100%;height:100%;object-fit:cover;display:block;transform:scaleX(-1)}
#cvs{position:absolute;top:0;left:0;width:100%;height:100%;pointer-events:none;z-index:3;transform:scaleX(-1)}
#fps{position:absolute;top:12px;right:12px;z-index:5;font-size:.63rem;color:var(--green);
  background:rgba(0,0,0,.65);padding:3px 8px;border:1px solid rgba(57,255,20,.25);border-radius:2px}
.live-tag{position:absolute;top:12px;left:12px;z-index:5;font-family:'Orbitron',sans-serif;
  font-size:.55rem;color:var(--accent2);background:rgba(0,0,0,.65);padding:3px 8px;
  border:1px solid rgba(255,0,110,.25);border-radius:2px;display:flex;align-items:center;gap:5px;letter-spacing:2px}
.rec{width:6px;height:6px;border-radius:50%;background:var(--accent2);animation:blink .8s infinite}

/* LOADING */
#loader{position:absolute;inset:0;background:rgba(3,8,15,.95);display:flex;flex-direction:column;
  align-items:center;justify-content:center;gap:20px;z-index:10}
.ld-title{font-family:'Orbitron',sans-serif;font-size:1rem;color:var(--accent);
  text-shadow:0 0 16px var(--accent);letter-spacing:4px}
.ld-track{width:280px;height:3px;background:rgba(0,245,255,.1);border-radius:2px;overflow:hidden;border:1px solid var(--border)}
.ld-bar{height:100%;background:var(--accent);box-shadow:0 0 8px var(--accent);width:0%;transition:width .4s}
.ld-msg{font-size:.65rem;color:var(--dim);text-align:center;max-width:280px;line-height:1.9}

/* BUTTONS */
.cam-btns{position:absolute;bottom:14px;left:50%;transform:translateX(-50%);z-index:5;display:flex;gap:10px}
.btn{padding:9px 20px;border:1px solid var(--accent);background:rgba(0,245,255,.07);color:var(--accent);
  font-family:'Orbitron',sans-serif;font-size:.6rem;font-weight:700;letter-spacing:2px;cursor:pointer;
  border-radius:2px;text-transform:uppercase;transition:.2s}
.btn:hover{background:rgba(0,245,255,.18);box-shadow:0 0 16px rgba(0,245,255,.3)}
.btn.red{border-color:var(--accent2);color:var(--accent2);background:rgba(255,0,110,.07)}
.btn.red:hover{background:rgba(255,0,110,.18);box-shadow:0 0 16px rgba(255,0,110,.3)}

/* SIDE PANELS */
.side{display:flex;flex-direction:column;gap:14px}
.panel{border:1px solid var(--border);border-radius:4px;background:var(--panel);overflow:hidden}
.ph{padding:9px 14px;background:rgba(0,245,255,.04);border-bottom:1px solid var(--border);
  font-family:'Orbitron',sans-serif;font-size:.6rem;font-weight:700;letter-spacing:3px;color:var(--accent)}

/* MOOD */
.mood-body{padding:16px;display:flex;flex-direction:column;align-items:center;gap:12px}
.mood-emoji{font-size:3.6rem;animation:float 3s ease-in-out infinite;filter:drop-shadow(0 0 12px currentColor)}
@keyframes float{0%,100%{transform:translateY(0)}50%{transform:translateY(-6px)}}
.mood-name{font-family:'Orbitron',sans-serif;font-size:1.15rem;font-weight:900;letter-spacing:3px;transition:.4s}
.mood-sub{font-size:.62rem;color:var(--dim);text-align:center;line-height:1.6;min-height:36px}
.mood-bars{width:100%}
.mb-row{display:flex;align-items:center;gap:8px;margin-bottom:5px;font-size:.64rem}
.mb-lbl{width:72px;color:var(--dim);text-transform:capitalize}
.mb-track{flex:1;height:5px;background:rgba(255,255,255,.05);border-radius:3px;overflow:hidden}
.mb-fill{height:100%;border-radius:3px;transition:width .35s ease}
.mb-pct{width:32px;text-align:right;font-size:.6rem}

/* STATS */
.sg{display:grid;grid-template-columns:1fr 1fr;gap:8px;padding:12px}
.sb{border:1px solid var(--border);border-radius:3px;padding:9px;text-align:center;background:rgba(0,245,255,.02)}
.sv{font-family:'Orbitron',sans-serif;font-size:1.3rem;font-weight:900;color:var(--accent);text-shadow:0 0 8px var(--accent)}
.sl{font-size:.58rem;color:var(--dim);margin-top:3px;letter-spacing:1px}

/* OBJECTS LIST */
.obj-list{padding:10px;overflow-y:auto;max-height:300px}
.oi{display:flex;align-items:center;gap:9px;padding:7px 9px;border:1px solid rgba(0,245,255,.08);
  border-radius:3px;margin-bottom:6px;background:rgba(0,245,255,.025);animation:fdin .25s ease}
@keyframes fdin{from{opacity:0;transform:translateX(8px)}to{opacity:1;transform:none}}
.oi:hover{border-color:var(--accent);background:rgba(0,245,255,.06)}
.oi-icon{font-size:1.4rem;flex-shrink:0}
.oi-info{flex:1;min-width:0}
.oi-name{font-family:'Orbitron',sans-serif;font-size:.67rem;font-weight:700;color:var(--accent);
  text-transform:uppercase;letter-spacing:1px}
.oi-desc{font-size:.58rem;color:var(--dim);margin-top:2px;line-height:1.4;
  white-space:nowrap;overflow:hidden;text-overflow:ellipsis}
.oi-conf{font-size:.62rem;color:var(--green);font-weight:700;text-align:right;flex-shrink:0}
.cb{width:38px;height:3px;background:rgba(57,255,20,.12);border-radius:2px;margin-top:3px;overflow:hidden}
.cf{height:100%;background:var(--green);border-radius:2px}
.empty{text-align:center;padding:24px 12px;color:var(--dim);font-size:.67rem;line-height:2}

/* TICKER */
.ticker{grid-column:1/3;border:1px solid var(--border);border-radius:4px;background:var(--panel);
  padding:9px 16px;display:flex;align-items:center;gap:12px;overflow:hidden}
.tk-lbl{font-family:'Orbitron',sans-serif;font-size:.58rem;color:var(--accent2);letter-spacing:2px;white-space:nowrap}
.tk-txt{font-size:.67rem;color:var(--text);white-space:nowrap;overflow:hidden;text-overflow:ellipsis}

::-webkit-scrollbar{width:3px}::-webkit-scrollbar-thumb{background:var(--border);border-radius:2px}
</style>
</head>
<body>

<header>
  <div class="logo">VISION<em>AI</em></div>
  <div class="hdr-status">
    <div class="st"><div class="dot" id="dc"></div><span id="lc">CAM OFFLINE</span></div>
    <div class="st"><div class="dot" id="dm"></div><span id="lm">LOADING MODEL</span></div>
    <div class="st"><div class="dot" id="dt"></div><span id="lt">TRACKING IDLE</span></div>
  </div>
</header>

<div class="layout">

  <!-- CAMERA -->
  <div class="cam-wrap" id="camWrap">
    <div class="corner tl"></div><div class="corner tr"></div>
    <div class="corner bl"></div><div class="corner br"></div>
    <div class="live-tag"><div class="rec"></div>LIVE</div>
    <div id="fps">-- FPS</div>
    <video id="vid" autoplay muted playsinline></video>
    <canvas id="cvs"></canvas>

    <div id="loader">
      <div class="ld-title">‚¨° LOADING VISION AI ‚¨°</div>
      <div class="ld-track"><div class="ld-bar" id="ldbar"></div></div>
      <div class="ld-msg" id="ldmsg">Downloading COCO-SSD neural network...<br>This takes ~10 seconds on first load</div>
    </div>

    <div class="cam-btns">
      <button class="btn" id="btnStart" onclick="startCam()">‚ñ∂ ACTIVATE CAMERA</button>
      <button class="btn red" id="btnStop" onclick="stopCam()" style="display:none">‚ñ† STOP</button>
    </div>
  </div>

  <!-- SIDE -->
  <div class="side">

    <!-- MOOD -->
    <div class="panel">
      <div class="ph">‚¨° MOOD ANALYSIS</div>
      <div class="mood-body">
        <div class="mood-emoji" id="mEmoji">üòê</div>
        <div class="mood-name" id="mName" style="color:var(--dim)">AWAITING FACE</div>
        <div class="mood-sub" id="mSub">Point your face at the camera to detect your emotion in real time.</div>
        <div class="mood-bars" id="mBars"></div>
      </div>
    </div>

    <!-- STATS -->
    <div class="panel">
      <div class="ph">‚¨° LIVE STATS</div>
      <div class="sg">
        <div class="sb"><div class="sv" id="sObj">0</div><div class="sl">OBJECTS</div></div>
        <div class="sb"><div class="sv" id="sFps">0</div><div class="sl">FPS</div></div>
        <div class="sb"><div class="sv" id="sFrames">0</div><div class="sl">FRAMES</div></div>
        <div class="sb"><div class="sv" id="sTracked">0</div><div class="sl">TRACKED</div></div>
      </div>
    </div>

    <!-- OBJECTS -->
    <div class="panel" style="flex:1">
      <div class="ph">‚¨° DETECTED OBJECTS</div>
      <div class="obj-list" id="objList">
        <div class="empty">Press ACTIVATE CAMERA<br>to start detection</div>
      </div>
    </div>

  </div>

  <!-- TICKER -->
  <div class="ticker">
    <span class="tk-lbl">‚ñ∂ INFO</span>
    <span class="tk-txt" id="tkTxt">VisionAI ready ‚Äî click Activate Camera to begin real-time detection and mood analysis.</span>
  </div>

</div>

<script src="https://cdn.jsdelivr.net/npm/@tensorflow/tfjs@4.10.0/dist/tf.min.js"></script>
<script src="https://cdn.jsdelivr.net/npm/@tensorflow-models/coco-ssd@2.2.2/dist/coco-ssd.min.js"></script>
<script>
// ‚îÄ‚îÄ OBJECT METADATA ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
const OBJ = {
  person:     {i:'üßë',d:'Human detected. Tracking body position.',c:'#00f5ff'},
  cat:        {i:'üê±',d:'Domestic cat. Felis catus.',c:'#ff9f43'},
  dog:        {i:'üê∂',d:'Domestic dog. Canis lupus familiaris.',c:'#ff9f43'},
  car:        {i:'üöó',d:'Automobile vehicle.',c:'#54a0ff'},
  truck:      {i:'üöõ',d:'Large motor truck.',c:'#54a0ff'},
  bus:        {i:'üöå',d:'Public transport bus.',c:'#54a0ff'},
  bicycle:    {i:'üö≤',d:'Human-powered two-wheel vehicle.',c:'#00b894'},
  motorcycle: {i:'üèçÔ∏è',d:'Two-wheeled motor vehicle.',c:'#00b894'},
  chair:      {i:'ü™ë',d:'Seating furniture.',c:'#a29bfe'},
  couch:      {i:'üõãÔ∏è',d:'Upholstered sofa for sitting.',c:'#a29bfe'},
  bottle:     {i:'üç∂',d:'Container for liquids.',c:'#55efc4'},
  cup:        {i:'‚òï',d:'Small cup for beverages.',c:'#ffeaa7'},
  bowl:       {i:'ü•£',d:'Round open container for food.',c:'#ffeaa7'},
  laptop:     {i:'üíª',d:'Portable personal computer.',c:'#74b9ff'},
  keyboard:   {i:'‚å®Ô∏è',d:'Input device for typing.',c:'#b2bec3'},
  mouse:      {i:'üñ±Ô∏è',d:'Computer pointing device.',c:'#b2bec3'},
  tv:         {i:'üì∫',d:'Television display screen.',c:'#6c5ce7'},
  cell_phone: {i:'üì±',d:'Mobile phone device.',c:'#fd79a8'},
  book:       {i:'üìö',d:'Printed or written document.',c:'#fdcb6e'},
  clock:      {i:'üïê',d:'Timekeeping device.',c:'#e17055'},
  scissors:   {i:'‚úÇÔ∏è',d:'Cutting tool with two blades.',c:'#636e72'},
  backpack:   {i:'üéí',d:'Bag worn on the back.',c:'#fdcb6e'},
  handbag:    {i:'üëú',d:'Small handheld bag.',c:'#fd79a8'},
  umbrella:   {i:'‚òÇÔ∏è',d:'Portable rain protection device.',c:'#74b9ff'},
  tie:        {i:'üëî',d:'Formal neckwear garment.',c:'#6c5ce7'},
  sports_ball:{i:'‚öΩ',d:'Spherical sports ball.',c:'#e17055'},
  kite:       {i:'ü™Å',d:'Flying object on a string.',c:'#00b894'},
  pizza:      {i:'üçï',d:'Italian flatbread dish.',c:'#ff9f43'},
  apple:      {i:'üçé',d:'Red fruit from apple tree.',c:'#e17055'},
  orange:     {i:'üçä',d:'Citrus fruit.',c:'#ff9f43'},
  banana:     {i:'üçå',d:'Yellow curved fruit.',c:'#ffeaa7'},
  sandwich:   {i:'ü•™',d:'Bread with filling.',c:'#fdcb6e'},
  cake:       {i:'üéÇ',d:'Baked sweet dessert.',c:'#fd79a8'},
  chair_bed:  {i:'üõèÔ∏è',d:'Furniture for sleeping.',c:'#a29bfe'},
  vase:       {i:'üè∫',d:'Decorative container.',c:'#55efc4'},
  toothbrush: {i:'ü™•',d:'Dental hygiene brush.',c:'#00b894'},
  _default:   {i:'üì¶',d:'Unclassified object tracked.',c:'#a29bfe'}
};

// ‚îÄ‚îÄ MOOD CONFIG ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
const MOODS = {
  happy:    {e:'üòÑ',c:'#f9ca24',d:'Smile and raised cheeks detected. Positive emotion!'},
  sad:      {e:'üò¢',c:'#74b9ff',d:'Downcast expression. Low energy facial cues.'},
  angry:    {e:'üò°',c:'#ff7675',d:'Tense jaw and furrowed brow detected.'},
  surprised:{e:'üò≤',c:'#fdcb6e',d:'Wide eyes and open mouth detected.'},
  fearful:  {e:'üò®',c:'#a29bfe',d:'Raised brow, widened eyes detected.'},
  neutral:  {e:'üòê',c:'#b2bec3',d:'Relaxed muscles. No dominant emotion.'},
  disgusted:{e:'ü§¢',c:'#55efc4',d:'Nose wrinkle, upper lip raised.'}
};

// ‚îÄ‚îÄ STATE ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
let model = null, stream = null, running = false;
let frameCount = 0, fps = 0, lastFpsT = Date.now(), totalFrames = 0, totalTracked = 0;
let trackerMap = {}, trackCnt = 0;
let detectEvery = 3, frameN = 0; // skip frames for speed
let lastPreds = [], lastMood = null;
let moodFrameCount = 0;
let raf = null;

// ‚îÄ‚îÄ MODEL LOAD ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
(async () => {
  const bar = id('ldbar'), msg = id('ldmsg');
  setDot('dm','warn'); setLabel('lm','LOADING...');
  bar.style.width = '15%';
  msg.innerHTML = 'Downloading COCO-SSD model (~5MB)...<br><span style="color:var(--warn)">Please wait, first load takes ~10s</span>';

  try {
    // Use backend=cpu for stability if webgl fails
    try { await tf.setBackend('webgl'); } catch(e) { await tf.setBackend('cpu'); }
    bar.style.width = '40%';
    msg.innerHTML = 'TensorFlow ready. Loading detection model...';

    model = await cocoSsd.load({ base: 'lite_mobilenet_v2' }); // fastest variant
    bar.style.width = '100%';
    msg.innerHTML = '‚úì Model loaded! Click Activate Camera.';
    setDot('dm','on'); setLabel('lm','MODEL READY');
    await sleep(600);
    id('loader').style.display = 'none';
  } catch(e) {
    bar.style.width='100%'; bar.style.background='var(--accent2)';
    msg.innerHTML = '‚úó Load failed: ' + e.message + '<br>Check internet connection.';
    setDot('dm','err'); setLabel('lm','LOAD ERROR');
  }
})();

// ‚îÄ‚îÄ CAMERA ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
async function startCam() {
  if (!model) { tick('‚ö† Model still loading, please wait a moment...'); return; }
  try {
    stream = await navigator.mediaDevices.getUserMedia({video:{width:640,height:480,facingMode:'user'},audio:false});
    const vid = id('vid'), cvs = id('cvs');
    vid.srcObject = stream;
    await vid.play();
    vid.addEventListener('loadedmetadata',()=>{
      cvs.width = vid.videoWidth; cvs.height = vid.videoHeight;
      running = true;
      setDot('dc','on'); setLabel('lc','CAM ACTIVE');
      setDot('dt','on'); setLabel('lt','TRACKING ON');
      id('btnStart').style.display='none'; id('btnStop').style.display='block';
      loop();
    });
  } catch(e) {
    setDot('dc','err'); setLabel('lc','CAM DENIED');
    tick('Camera access denied. Allow camera in browser settings, then retry.');
  }
}

function stopCam() {
  running = false;
  if (raf) cancelAnimationFrame(raf);
  if (stream) stream.getTracks().forEach(t=>t.stop());
  id('vid').srcObject = null;
  const cvs = id('cvs'); cvs.getContext('2d').clearRect(0,0,cvs.width,cvs.height);
  setDot('dc',''); setLabel('lc','CAM OFFLINE');
  setDot('dt',''); setLabel('lt','TRACKING IDLE');
  id('btnStart').style.display='block'; id('btnStop').style.display='none';
  id('objList').innerHTML='<div class="empty">Camera stopped.</div>';
  moodIdle(); trackerMap={}; lastPreds=[];
  id('sObj').textContent='0';
}

// ‚îÄ‚îÄ MAIN LOOP ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
async function loop() {
  if (!running) return;

  const vid = id('vid'), cvs = id('cvs'), ctx = cvs.getContext('2d');

  // FPS counter
  frameCount++; totalFrames++;
  const now = Date.now();
  if (now - lastFpsT >= 1000) {
    fps = frameCount; frameCount = 0; lastFpsT = now;
    id('fps').textContent = fps + ' FPS';
    id('sFps').textContent = fps;
  }

  // Run detection every N frames to keep it smooth
  frameN++;
  if (frameN % detectEvery === 0 && vid.readyState === 4) {
    try {
      lastPreds = await model.detect(vid, 10); // max 10 detections
    } catch(e) { lastPreds = []; }
    updateTracker(lastPreds);
    updateObjectsUI(lastPreds);
    id('sObj').textContent = lastPreds.length;
    id('sFrames').textContent = totalFrames;
    id('sTracked').textContent = totalTracked;

    // Mood from pixel analysis every 10 frames
    if (frameN % 10 === 0) analyzeMoodFromPixels(vid);
  }

  // Draw every frame (smooth)
  ctx.clearRect(0,0,cvs.width,cvs.height);
  drawDetections(ctx, lastPreds, cvs.width, cvs.height);

  raf = requestAnimationFrame(loop);
}

// ‚îÄ‚îÄ TRACKER ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
function updateTracker(preds) {
  const next = {};
  for (const p of preds) {
    const [x,y,w,h] = p.bbox;
    const cx = x+w/2, cy = y+h/2;
    let bestId = null, bestDist = 100;
    for (const [id2, obj] of Object.entries(trackerMap)) {
      if (obj.cls !== p.class) continue;
      const d = Math.hypot(cx-obj.cx, cy-obj.cy);
      if (d < bestDist) { bestDist=d; bestId=id2; }
    }
    if (bestId) {
      next[bestId] = {cls:p.class, cx, cy, age:(trackerMap[bestId].age||0)+1};
    } else {
      const nid = ++trackCnt; totalTracked++;
      next[nid] = {cls:p.class, cx, cy, age:1};
    }
    p._tid = bestId || trackCnt;
  }
  trackerMap = next;
}

// ‚îÄ‚îÄ DRAW ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
function drawDetections(ctx, preds, W, H) {
  for (const p of preds) {
    const info = OBJ[p.class] || OBJ._default;
    let [x,y,w,h] = p.bbox;
    // clamp
    x=Math.max(0,x); y=Math.max(0,y);
    w=Math.min(w,W-x); h=Math.min(h,H-y);
    const conf = Math.round(p.score*100);
    const col = info.c;
    const tid = p._tid || '';

    // Glow box
    ctx.shadowColor=col; ctx.shadowBlur=10;
    ctx.strokeStyle=col; ctx.lineWidth=1.5;
    ctx.strokeRect(x,y,w,h);
    ctx.shadowBlur=0;

    // Corner marks
    const cs=10; ctx.strokeStyle='#fff'; ctx.lineWidth=2;
    for(const[cx2,cy2,dx,dy] of [[x,y,1,1],[x+w,y,-1,1],[x,y+h,1,-1],[x+w,y+h,-1,-1]]){
      ctx.beginPath();ctx.moveTo(cx2+dx*cs,cy2);ctx.lineTo(cx2,cy2);ctx.lineTo(cx2,cy2+dy*cs);ctx.stroke();
    }

    // Label
    const lbl = `${p.class.replace('_',' ').toUpperCase()}  #${tid}  ${conf}%`;
    ctx.font = 'bold 11px "Share Tech Mono",monospace';
    const tw = ctx.measureText(lbl).width;
    const lx = x, ly = y>20 ? y-20 : y+h+2;
    ctx.fillStyle='rgba(0,0,0,.78)';
    ctx.fillRect(lx,ly,tw+12,17);
    ctx.fillStyle=col;
    ctx.fillText(lbl,lx+5,ly+12);

    // Confidence bar at bottom of box
    ctx.fillStyle='rgba(0,0,0,.45)'; ctx.fillRect(x,y+h-4,w,4);
    ctx.fillStyle=col; ctx.fillRect(x,y+h-4,w*p.score,4);
  }
}

// ‚îÄ‚îÄ MOOD FROM PIXEL ANALYSIS ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
// Uses a hidden canvas to sample face-region pixels and infer rough mood
// Not a real model ‚Äî uses brightness, contrast, movement heuristics for fun
let prevBrightness = 0, moodHistory = [];
function analyzeMoodFromPixels(vid) {
  try {
    const tmpC = document.createElement('canvas');
    tmpC.width=80; tmpC.height=60;
    const tc = tmpC.getContext('2d');
    tc.drawImage(vid,0,0,80,60);
    const px = tc.getImageData(0,0,80,60).data;

    let r=0,g=0,b=0,n=px.length/4;
    for(let i=0;i<px.length;i+=4){r+=px[i];g+=px[i+1];b+=px[i+2];}
    r/=n; g/=n; b/=n;
    const brightness = (r+g+b)/3;
    const warmth = r - b; // warm=positive, cool=negative
    const delta = Math.abs(brightness - prevBrightness);
    prevBrightness = brightness;

    // Heuristic mood scores (fun approximation)
    let scores = {
      happy: 0, sad: 0, angry: 0, surprised: 0,
      fearful: 0, neutral: 0, disgusted: 0
    };

    // Brighter frame ‚Üí more positive
    if (brightness > 160) scores.happy += 0.4;
    else if (brightness < 80) scores.sad += 0.35;

    // Warm tones ‚Üí happy/angry
    if (warmth > 20) scores.happy += 0.25;
    else if (warmth < -15) scores.sad += 0.2;

    // High movement ‚Üí surprised/fearful
    if (delta > 18) scores.surprised += 0.4;
    else if (delta > 10) scores.fearful += 0.2;

    // Normalize & add neutral
    const total = Object.values(scores).reduce((a,b)=>a+b,0);
    if (total < 0.3) scores.neutral = 0.7;
    else scores.neutral += Math.max(0, 0.3 - total*0.1);

    // Smooth with history
    moodHistory.push(scores);
    if (moodHistory.length > 5) moodHistory.shift();
    const avg = {};
    for (const k of Object.keys(scores)) {
      avg[k] = moodHistory.reduce((s,m)=>s+(m[k]||0),0)/moodHistory.length;
    }

    // Normalize
    const sum2 = Object.values(avg).reduce((a,b)=>a+b,0)||1;
    for(const k of Object.keys(avg)) avg[k] = avg[k]/sum2;

    updateMoodUI(avg);
  } catch(e) {}
}

function updateMoodUI(scores) {
  const sorted = Object.entries(scores).sort((a,b)=>b[1]-a[1]);
  const [topName, topVal] = sorted[0];
  const cfg = MOODS[topName] || MOODS.neutral;

  id('mEmoji').textContent = cfg.e;
  id('mName').textContent = topName.toUpperCase();
  id('mName').style.color = cfg.c;
  id('mSub').textContent = cfg.d;

  id('mBars').innerHTML = sorted.map(([n,v])=>{
    const mc = (MOODS[n]||MOODS.neutral).c;
    return `<div class="mb-row">
      <div class="mb-lbl">${n}</div>
      <div class="mb-track"><div class="mb-fill" style="width:${Math.round(v*100)}%;background:${mc}"></div></div>
      <div class="mb-pct" style="color:${mc}">${Math.round(v*100)}%</div>
    </div>`;
  }).join('');

  tick(`Mood: ${topName.toUpperCase()} (${Math.round(topVal*100)}%) ‚Äî ${cfg.d}`);
}

function moodIdle() {
  id('mEmoji').textContent='üòê';
  id('mName').textContent='NO FEED';
  id('mName').style.color='var(--dim)';
  id('mSub').textContent='Activate camera to analyze your mood in real time.';
  id('mBars').innerHTML='';
}

// ‚îÄ‚îÄ OBJECTS LIST UI ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
function updateObjectsUI(preds) {
  const el = id('objList');
  if (!preds.length) {
    el.innerHTML='<div class="empty">No objects detected.<br>Move something into frame!</div>';
    return;
  }
  // Deduplicate by class, keep highest confidence
  const uniq = {};
  for(const p of preds) if(!uniq[p.class]||p.score>uniq[p.class].score) uniq[p.class]=p;

  el.innerHTML = Object.values(uniq).sort((a,b)=>b.score-a.score).map(p=>{
    const info = OBJ[p.class]||OBJ._default;
    const conf = Math.round(p.score*100);
    return `<div class="oi">
      <div class="oi-icon">${info.i}</div>
      <div class="oi-info">
        <div class="oi-name">${p.class.replace(/_/g,' ')}</div>
        <div class="oi-desc">${info.d}</div>
      </div>
      <div style="flex-shrink:0;text-align:right">
        <div class="oi-conf">${conf}%</div>
        <div class="cb"><div class="cf" style="width:${conf}%"></div></div>
      </div>
    </div>`;
  }).join('');

  const names = Object.keys(uniq).map(n=>n.replace(/_/g,' ').toUpperCase()).join(', ');
  tick(`Detected: ${names} ‚Äî ${preds.length} object(s) actively tracked.`);
}

// ‚îÄ‚îÄ HELPERS ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
const id = s => document.getElementById(s);
const sleep = ms => new Promise(r=>setTimeout(r,ms));
function tick(msg) { id('tkTxt').textContent = msg; }
function setDot(elId, state) {
  const d = id(elId); d.className='dot';
  if(state==='on') d.className='dot on';
  else if(state==='err') d.className='dot err';
  else if(state==='warn') d.className='dot warn';
}
function setLabel(elId, txt) { id(elId).textContent = txt; }
</script>
</body>
</html>
```
