Step 1: Import Libraries

```python
import numpy as np              # Math & arrays ke liye
import matplotlib.pyplot as plt # Graph plotting ke liye
```

Step 2: Create Dummy Data

```python
# Exercise time (hours)
data = np.array([1, 2, 3, 4, 5])

# True calories burned (target)
true_calories = np.array([200, 400, 600, 800, 1000])
```

Step 3: Loss Function (Mean Squared Error)

```python
def calculate_loss(weight):
    # Prediction = weight * time
    predictions = weight * data

    # Mean Squared Error
    loss = np.mean((predictions - true_calories) ** 2)
    return loss
```

Step 4: Gradient Function

```python
def calculate_gradient(data, weight):
    # Prediction
    predictions = weight * data

    # Gradient of MSE with respect to weight
    gradient = np.mean(2 * data * (predictions - true_calories))
    return gradient
```

Task 2: Testing the “Step Size” (Learning Rate)

We test two learning rates:

0.000001 → very slow

0.1 → fast convergenc

Step 5: Gradient Descent Function (FIXED)

```python
def train_model(learning_rate):
    weight = 0.5                 # Initial weight
    losses = []                  # Loss history store karne ke liye

    for i in range(100):
        gradient = calculate_gradient(data, weight)

        # ✅ CORRECT UPDATE RULE (minus sign)
        weight = weight - (learning_rate * gradient)

        # Loss calculate karna
        loss = calculate_loss(weight)
        losses.append(loss)

    return losses
```

Step 6: Run Training for Both Learning Rates

```python
# Very small learning rate
loss_slow = train_model(0.000001)

# Larger learning rate
loss_fast = train_model(0.1)
```

Step 7: Plot Loss Curves (Required)

```python
plt.plot(loss_slow, label="Learning Rate = 0.000001")
plt.plot(loss_fast, label="Learning Rate = 0.1")

plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.title("Loss Curve Comparison")
plt.legend()

plt.show()
```
